```python
'''
Q3. What are the diffent ways to define and call a function in pyton?
Answer:
    There are different types to define and call a function, which makes your codes more flexible.
    Different ways to define and call function are as following:

1.Simple Function Defination and Call:
    Defination: An basic function which does not take any parameters while creating.
    Call: You can call function directly in this method.
Example:
'''
# Creating a function
def greet():
    return "Hi how are you"
# Calling function
print(greet())
'''
2.Function with Parameters:
    Definition: Functions are given some parameters, which you can pass the value, while calling the function.
    Call: Function are called with the parameter.
Example:
'''
def greeting(name):
    return "hi" , name
print(greeting("asad"))

'''
3. Functions with Default Parameters:
    Definition: Function is created with a default parameters, and it is used when no arguement was passed while calling the function.
    Call: You can call a function with default value or without a default value
Example:
'''
def greet(name = "asad"): # default value (in paramter) given to a function
    return "hello", name

print(greet("Asad")) # calling wiht arguement
#Output will be hello Asad

print(greet()) # With no arguement (default used: "asad")
#Output will be hello asad
'''
4.Function with Variable-Length Arguements:
    Definition: If you waant to pass multiple arguements in a single function and you dont know the exact numbers of the parameter,
        in this case you can use **kwargs keyword in the parameter.
    Call: You can pass multiple Arguements.
Example:
'''
def add_numbers(*args): # takes multiple positional arguements
    return sum(args)

print(add_numbers(1, 2, 3, 4, ))# calls with variable number of arguements
# OUtput will be: 10

'''
5.Lambda Functions: (anonymous function)
    Definition: Lambda is an anonymous function, which is defined in a single line of code, 
        Lambda function is use for small operations.
    Call: You can call lambda function, directly or with a variable.
Example:
'''
multiply = lambda x, y: x*y # Lambda function
print(multiply(6, 9)) # function call
# OUtput will be: 54
'''
6.Recursive Function:
    Definition: An function which calls himself(function itself) whitin the function is called Recursive function,
    this usually to divide a problem.
    Call: You can recursive through calling it in the base case.
Example:
'''
def recursive_sum(n):
    # Base case: when n is 0, it will return 0
    if n == 0:
        return 0
    # Recursive case: call n on every call, for (n-1) call function call
    else:
        return n + recursive_sum(n - 1)

# Example call
result = recursive_sum(5)
print(result)  # Output will be 15 (5 + 4 + 3 + 2 + 1 + 0)
print()
'''
7.Function inside of an function:
    Definition: You can define a function, inside of an another function.
    Call: Nested function is called inside of the outer function.
Exmaple:
'''
def outer_function():
    def inner_function():
        return "hello from inside function(nested)"
    return inner_function()
print(outer_function()) # calling function
# Output will be from inner function: "hello from inside function(nested)"
'''
Insight:
    There are different types to call a function, in which there are simple function, variable-length function, arguements function,
    lambdas and recursion function.
    This flexibility gives you the chance of code reusablility and modular, which works in every senario. 
'''
print()
```

```python
'''
# Q7. What are the advantages of using generator over regular functions?
Answer
    Generator offers several advantages over regular function, particularly when dean dealing with large datasets or continiously streams of data,
    rather than regular function that returns all results at once and consume memory for all the entire result set,
    generator prodece values ondemand, using the yield keyword.

Advantages if using Generators over regular function:
    1. Memory efficient(on demand generation)
        Generators only produce one value at a time, which means they dont need to hold the entire sequence in memory.
Example:
'''
def my_generator():
    for i in range (1, 1000000):
        yield i
'''
here my_generator is generating a large sequence of value, but it will only product(generate) one at a time,
regular function there, so it will all the values is stored in the memory at a time, 
so it will consume time to return and too stores in the memory , which not use the memory efficiently 

    *.Lazy Evaluation:
        Generators follow lazy evaluation, meaning values are prduced only when needed, 
        this reduces overhead when not all values are required immediately.
Example:
'''
def lazy_number():
    yield 1
    yield 2 #iterator that uses yield keyword
    yield 3
gen = lazy_number()
print(next(gen)) #output: 1
# in this as you generate thee next value using __next__() function, so you dont store all the values immediately in the memory.
'''
    *.State Retention without recursion overhead:
        Generators maintain their state between yield call wihtout the need for recursion,
        avoiding the deep recursion issues that regular functions might face.
Example
'''
def count_down(num):
    while num > 0:
        yield num
        num -= 1
gen = count_down(5)
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
# print(next(gen)) # this will pproduce error because elements are over (iterating)
'''
Insights:
    Generators provide a highly memory efficient and lazy way to handle large datasets or infinite sequences,
    instead of loading the entire datasets into memory at once, they generates the value (ondemand),
    reducing memory usage and making your code more performance efficient for real time data or for streams , 
    generators become an essential tool, as they dynamically produce results only when required.
    '''
print()
```

```python
'''
Q8.What is a lambda function in python and when it is typically used?
Answer:
    Lambda is a anonymours(nameless) function in python, which is used to express small function and it is defined in a single line,
    and does not need return statement to return the value, because whichever function you write in lambda, it automatically returns the value.

Lambda function Definition: 
        syntax:
    lambda arguements: expression
#A lambda function can take multiple arguements, but it always returns the result of single expression.

UseCase:
    One time use:
        Lambda function are often used when you need a function for a single use or temporary purpose .
Example:
'''
add = lambda x, y: x+y
print(add(4, 4))
#Here, the lambda function adds two arguement without any needing a formal function definition.
'''
1.Used as arguements in function:
    lambda frequently used as arguements to other functions like map(), filter() and sorted().
Example:
'''
numbers = [1, 3, 5, 7, 9]
squared_numbers = list(map(lambda x: x**2, numbers))
print(squared_numbers)
# In this example the lambda function is used within map() to square each number in the list.

'''
For custom sorting or Key functions:
    lambda functions are often used in sorting or when a custom key function is needed in functions like sorted()
Example:
'''
students = [("asad", 99), ("pate",98), ("ahmed",97)]
sorted_students = sorted(students, key=lambda x: x[1])
print(sorted_students)
#Here, the lambda function sorts a list of tuples based on the second element(maeks).
'''
insights:
    Lambda functions make you code more concise and efficient, when you need a small, one time function.
    They are ideal for situation where a full function definition would be unnecessary, by using lambda function, 
    you can quickly perform single operations, specially in case involving real time data processing or when passing functions as arguements.
    '''
print()
```

```python
'''
Q9. Explain the purpose and usage of the map() function in python?
Answer:
    The main purpose of map() is to transform data by applying a function to each element of the sequence,
    without using explicit loops, which helps make the code more concise and readable.

Usage of map()
        Syntax
    map(function, iterable)

1.Arguements:
    *function: Function that you want to apply to each item of the iterable.
    *iterable: A sequence of like(tuple, list, string) whose elements will be passed to the function.
2.Return value:
    The result is a map object (which is an iterator), 
    you can convert this into a list or tuple, if you need to view or store the the result.
Example:
    1.Basic Example: Applying a function to square each number in a list.
'''
def square(x):
    return x ** 2
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(square, numbers))
print(squared_numbers)
'''
    2.Using Lambda with map(): you can also use a lambda function directly whitin map() to avoid writting seperate functions.
Example:
'''
numbers = [1, 2, 3, 4, 5]
square_numbers = list(map(lambda x,: x**2, numbers))
print(square_numbers)
'''
    3. Working with multiple iterables: map() can also work with multiple iterables,
    if the function accepts more than one arguement.
Examples:
'''
numbers1 = [5, 6, 7, 8, 9]
numbers2 = [1, 2, 3, 4, 5]
result = list(map(lambda x, y: x + y, numbers1, numbers2))
print(result)
'''
insights:
    The map() is a powerful tool that allows you to apply transformations across a sequence of data,
    without explicitily writting a loop, it makes the code cleaner and is particularly useful for data processing or,
    when you need to apply operations across every element efficiently.
    '''
print()
```

```python
'''
Q10. What is the difference between map(), reduce() and filter() functions in python
Answer:
    The difference between __map__(), __reduct__() and __filter__() are as following:
1.Map()
    When you want to apply a function to every item in ana iterable.
        Map applies a given function to each item of an iterable, for eg, list, tuple and returns a new iterable with the result.
    UseCase of map : we use it when, we want to perform all elements in an iterable.

        map Syntax:
    map(function, iterable)

Example:
'''
numbers = [6, 7, 8 , 9 ]
result = list(map(lambda x : x * 2 , numbers))
print(result)
# This function, map is iteration on each element and implement the lambda function on each element to multiply every element by 2

'''
2. Filter()
    When you want to select items based on a condition.
    Transforms element, and returns iterable like list, etc
        It filters the element and returns iterables with selected items.
        Filter() function applies a function that returns a boolean (True / False) to each item an iterable,
        and only the item that return True are included in the output.
    UseCase of filter: Use it when you want to filter out certain elements based on a condition.

        filter Syntax:
    filter(function, iterable) 
Example:
'''
number = [1, 2, 3 ,4, 5, 7]
final_output = list(filter(lambda x: x % 2 == 0, number))
print(final_output)
# This pragram takes a number list, filter will iterate every element in the list 
# and whichever elements is divided by 2 and the remainder of that element will be 0, it will print that number.

'''
3. Reduce()
    When you want to aggregate(single ouput ) or reduce all elements to a single element.
    Reduce purpose is to reduce elemetns , and it return a single value
        Reduce is not a inbuit function in python, you have to import if using(import key word)
        Reduce use a function that takes two arguement aggregate to the items of an itearable, reducing it to a single value.
    Usage: When you need to combine or reduct the element to a single result for eg sum, multiply, etc

        reduce Syntax:
    reduce(function, iterable)

Example:
'''
from functools import reduce
num = [1, 3 ,4 ,5, 6 ,7]
adding = reduce(lambda x, y: x + y, num)
print(adding)
# In this reduce funtion we are combining(sum) the numbers in the list using lambda 



'''
Insights:
    These functions are essential for functional programming in python, making code clearner and more efficient .
    Map is ideal for transforming data, filter for selection subsets of data base on condition, and reduce for combining all elements into one result
    These are especially useful when working with large datasets, providing clear and concise solutions to commom problems
'''

print()
```
